```python
# from statsmodels.tsa.stattools import adfuller

# # Function to perform and print ADF test results
# def run_adf_test(series, title='Augmented Dickey-Fuller Test'):
#     print(f"--- {title} ---")
#     result = adfuller(series, autolag='AIC')
#     labels = ['ADF Statistic', 'p-value', 'Lags Used', 'Number of Observations Used']
#     for value, label in zip(result[0:4], labels):
#         print(f'{label}: {value:.4f}')
#     print('Critical Values:')
#     for key, value in result[4].items():
#         print(f'   {key}: {value:.4f}')
#     if result[1] <= 0.05:
#         print("Conclusion: The series is likely stationary (p-value <= 0.05).")
#     else:
#         print("Conclusion: The series is likely non-stationary (p-value > 0.05).")
#     print("\n")
#     return result[1]

# # Perform ADF test on original 'Close Price' series
# p_value_original = run_adf_test(df['Close Price'], title='ADF Test on Original Close Price')

# # If non-stationary, perform first-order differencing and re-test
# if p_value_original > 0.05:
#     print("Original 'Close Price' is non-stationary. Applying first-order differencing...")
#     # Ensure the index is unique before creating the differenced series
#     df_unique_index = df[~df.index.duplicated(keep='first')]
#     df_unique_index['Close Price Diff'] = df_unique_index['Close Price'].diff().dropna()
#     run_adf_test(df_unique_index['Close Price Diff'], title='ADF Test on First-Order Differenced Close Price')
# else:
#     print("Original 'Close Price' is already stationary, no differencing needed.")









from statsmodels.tsa.stattools import adfuller

# Function to perform and print ADF test results
def run_adf_test(series, title='Augmented Dickey-Fuller Test'):
    print(f"--- {title} ---")
    series = series.dropna()
    result = adfuller(series, autolag='AIC')

    labels = ['ADF Statistic', 'p-value', 'Lags Used', 'Number of Observations Used']
    for value, label in zip(result[0:4], labels):
        print(f'{label}: {value:.4f}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value:.4f}')
    if result[1] <= 0.05:
        print("Conclusion: The series is likely stationary (p-value <= 0.05).")
    else:
        print("Conclusion: The series is likely non-stationary (p-value > 0.05).")
    print("\n")
    return result[1]


# Perform ADF test on original 'Close Price' series
p_value_original = run_adf_test(df['Close Price'], title='ADF Test on Original Close Price')


# If non-stationary, perform first-order differencing and re-test
if p_value_original > 0.05:
    print("Original 'Close Price' is non-stationary. Applying first-order differencing...")

    # FIX 2: Correct way to deduplicate index
    df_unique_index = df.loc[~df.index.duplicated(keep='first')].copy()

    # FIX 3: No dropna while assigning
    df_unique_index['Close Price Diff'] = df_unique_index['Close Price'].diff()

    # Re-test
    run_adf_test(df_unique_index['Close Price Diff'], title='ADF Test on First-Order Differenced Close Price')

else:
    print("Original 'Close Price' is already stationary, no differencing needed.")
```

**Reasoning**:
The `MissingDataError: exog contains inf or nans` indicates that the series passed to the `adfuller` function contains `NaN` or `inf` values. Although `dropna()` was used on the differenced series, it might not have caught all edge cases or `inf` values. To resolve this, I will explicitly replace `inf` values with `NaN` and then drop all `NaN` values from the series within the `run_adf_test` function, ensuring a clean series is passed to `adfuller`.



```python
from statsmodels.tsa.stattools import adfuller
import numpy as np

# Function to perform and print ADF test results
def run_adf_test(series, title='Augmented Dickey-Fuller Test'):
    print(f"--- {title} ---")
    # Explicitly replace inf values with NaN and then drop all NaN values
    cleaned_series = series.replace([np.inf, -np.inf], np.nan).dropna()

    if cleaned_series.empty:
        print("Error: Series is empty after cleaning. Cannot perform ADF test.")
        return 1.0 # Return a p-value indicating non-stationarity
    if len(cleaned_series) < 2: # ADF test requires at least 2 observations
        print("Error: Series has too few observations after cleaning. Cannot perform ADF test.")
        return 1.0

    result = adfuller(cleaned_series, autolag='AIC')
    labels = ['ADF Statistic', 'p-value', 'Lags Used', 'Number of Observations Used']
    for value, label in zip(result[0:4], labels):
        print(f'{label}: {value:.4f}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value:.4f}')
    if result[1] <= 0.05:
        print("Conclusion: The series is likely stationary (p-value <= 0.05).")
    else:
        print("Conclusion: The series is likely non-stationary (p-value > 0.05).")
    print("\n")
    return result[1]

# Perform ADF test on original 'Close Price' series
p_value_original = run_adf_test(df['Close Price'], title='ADF Test on Original Close Price')

# If non-stationary, perform first-order differencing and re-test
if p_value_original > 0.05:
    print("Original 'Close Price' is non-stationary. Applying first-order differencing...")
    # Ensure the index is unique before creating the differenced series and make a copy
    df_cleaned = df[~df.index.duplicated(keep='first')].copy()

    # Calculate differenced series, explicitly dropna initially
    differenced_series = df_cleaned['Close Price'].diff()

    # Pass the potentially differenced series to run_adf_test, which will perform final cleaning
    run_adf_test(differenced_series, title='ADF Test on First-Order Differenced Close Price')
else:
    print("Original 'Close Price' is already stationary, no differencing needed.")
```

## Generate PDF Histogram with Overlayed Distributions

### Subtask:
Create a histogram of the 'Close Price' to represent its probability density function. On the same graph, fit and plot the probability density functions for various distributions including beta, chi-squared, exponential, gamma, normal, Rayleigh, and Weibull, to visually compare their fit to the data. Ensure all plots are clearly labeled and include a legend.


**Reasoning**:
To visualize the distribution of 'Close Price' and compare it against various theoretical distributions, I will create a histogram of the data and then overlay the fitted Probability Density Functions (PDFs) for beta, chi-squared, exponential, gamma, normal, Rayleigh, and Weibull distributions.



```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

# Ensure close_price_data is available. It was created in the previous step.
# close_price_data = df_cleaned['Close Price']

# Create a figure and an axes object for plotting
plt.figure(figsize=(15, 8))
ax = plt.gca()

# Plot a histogram of the close_price_data with normalized density
ax.hist(close_price_data, bins=50, density=True, alpha=0.6, color='skyblue', label='Close Price Histogram')

# Define a list of distributions to fit and plot
distributions = [
    stats.beta,
    stats.chi2,
    stats.expon,
    stats.gamma,
    stats.norm,
    stats.rayleigh,
    stats.weibull_min
]

# Generate x-values for plotting the PDF, spanning the range of close_price_data
x_values = np.linspace(close_price_data.min(), close_price_data.max(), 500)

# Dictionary to store fitted parameters
fitted_params_dict = {}

# Iterate through each distribution, fit it, and plot its PDF
for dist in distributions:
    try:
        # Fit the distribution to close_price_data
        params = dist.fit(close_price_data)
        fitted_params_dict[dist.name] = params

        # Calculate the PDF values for the fitted distribution
        pdf_values = dist.pdf(x_values, *params)

        # Plot the PDF curve
        ax.plot(x_values, pdf_values, label=f'{dist.name.capitalize()} PDF')
    except Exception as e:
        print(f"Could not fit {dist.name}: {e}")

# Set the title and labels
ax.set_title('Distribution of Close Price with Fitted PDFs')
ax.set_xlabel('Close Price')
ax.set_ylabel('Density')

# Add a legend
ax.legend()

# Display the plot
plt.grid(True)
plt.show()
```
